```python
n_goods=3

def write_to_file(filename):
    with open(filename, 'w') as f:
        f.write('model;\n')
        f.write(f'\n')
        
        f.write('# exp_s_0=(1-\n')
        for i in range(1, n_goods):
            f.write(f'exp(s_{i})')
            if i < n_goods-1:
                f.write(f'-')
            f.write('\n')    
        f.write(');\n')
        f.write(f'\n')
        
        f.write('# exp_s_0p=(1-\n')
        for i in range(1, n_goods):
            f.write(f'exp(s_{i}(+1))')
            if i < n_goods-1:
                f.write(f'-')
            f.write('\n')    
        f.write(');\n')
        f.write(f'\n')
        
        
        f.write('# exp_n_0=(1-\n')
        for i in range(1, n_goods):
            f.write(f'exp(n_{i})')
            if i < n_goods-1:
                f.write(f'-')
            f.write('\n')    
        f.write(');\n')
        f.write(f'\n')
        
        f.write('# exp_n_0p=(1-\n')
        for i in range(1, n_goods):
            f.write(f'exp(n_{i}(+1))')
            if i < n_goods-1:
                f.write(f'-')
            f.write('\n')    
        f.write(');\n')
        f.write(f'\n')
        
        f.write('# exp_n_0m=(1-\n')
        for i in range(1, n_goods):
            f.write(f'exp(n_{i}(-1))')
            if i < n_goods-1:
                f.write(f'-')
            f.write('\n')    
        f.write(');\n')
        f.write(f'\n')
        
        
        
        
        f.write('# ph=(\n')
        for i in range(0, n_goods):
            f.write(f'om_{i}*exp(c_{i})^((eps-1)/eps)')
            if i < n_goods-1:
                f.write(f'+')
            f.write('\n')    
        f.write(')^((1-tht*eps)/(eps-1));\n')
        f.write(f'\n')
        f.write('# ph_plus=(\n')
        for i in range(0, n_goods):
            f.write(f'om_{i}*exp(c_{i}(+1))^((eps-1)/eps)')
            if i < n_goods-1:
                f.write(f'+')
            f.write('\n')    
        f.write(')^((1-tht*eps)/(eps-1));\n')
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# v_{i}=om_{i}*exp(c_{i})^(-1/eps)*ph;\n')
       
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# v_plus{i}=om_{i}*exp(c_{i}(+1))^(-1/eps)*ph_plus;\n')
        
        f.write(f'\n')
        f.write(f'# q_0=(1-psi_n/2*(exp_n_0/exp_n_0m-1)^2);\n')
        for i in range(1, n_goods):
            f.write(f'# q_{i}=(1-psi_n/2*(exp(n_{i})/exp(n_{i}(-1))-1)^2);\n')
        
        f.write(f'\n')
        f.write(f'# q_plus0=(1-psi_n/2*(exp_n_0p/exp_n_0-1)^2);\n')
        for i in range(1, n_goods):
            f.write(f'# q_plus{i}=(1-psi_n/2*(exp(n_{i}(+1))/exp(n_{i})-1)^2);\n')
        
        f.write(f'\n')
        f.write(f'# fn_0=exp(a_0)*(1-alf)*(exp_s_0*exp(k(-1))/exp_n_0/q_0)^alf;\n')  
        for i in range(1, n_goods):
            f.write(f'# fn_{i}=exp(a_{i})*(1-alf)*(exp(s_{i})*exp(k(-1))/exp(n_{i})/q_{i})^alf;\n')   
            
        f.write(f'\n')
        f.write(f'# fn_plus0=exp(a_0(+1))*(1-alf)*(exp_s_0p*exp(k)/exp_n_0p/q_plus0)^alf;\n')   
        for i in range(1, n_goods):
            f.write(f'# fn_plus{i}=exp(a_{i}(+1))*(1-alf)*(exp(s_{i}(+1))*exp(k)/exp(n_{i}(+1))/q_plus{i})^alf;\n')       
         

        f.write(f'\n')
        f.write(f'# fs_0=exp(a_0)*alf*(exp_n_0/exp_s_0*q_0)^(1-alf)*exp(k(-1))^alf;\n')
        for i in range(1, n_goods):
            f.write(f'# fs_{i}=exp(a_{i})*alf*(exp(n_{i})/exp(s_{i})*q_{i})^(1-alf)*exp(k(-1))^alf;\n')
        f.write(f'# fs_0plus=exp(a_0(+1))*alf*(exp_n_0p/exp_s_0p*q_plus0)^(1-alf)*exp(k)^alf;\n')

        
        f.write(f'\n')
        f.write(f'# x_10=(q_0-exp_n_0*psi_n/exp_n_0m*(exp_n_0/exp_n_0m-1));\n')
        for i in range(1, n_goods):
            f.write(f'# x_1{i}=(q_{i}-exp(n_{i})*psi_n/exp(n_{i}(-1))*(exp(n_{i})/exp(n_{i}(-1))-1));\n')
        
        f.write(f'\n')
        f.write(f'# x_20plus=psi_n*(exp_n_0p/exp_n_0-1)*(exp_n_0p/exp_n_0)^2;\n')
        for i in range(1, n_goods):
            f.write(f'# x_2{i}plus=psi_n*(exp(n_{i}(+1))/exp(n_{i})-1)*(exp(n_{i}(+1))/exp(n_{i}))^2;\n')
            
        f.write(f'\n')
        f.write(f'# z_1=1-psi_k*(i/exp(k(-1))-delt);\n')
        f.write(f'# z_1plus=1-psi_k*(i(+1)/exp(k)-delt);\n')       
            
        f.write(f'\n')
        f.write(f'# z_2plus=1-delt-psi_k/2*(i(+1)/exp(k)-delt)^2+psi_k*i(+1)/exp(k)*(i(+1)/exp(k)-delt);\n') 
        
           
        f.write(f'\n')
        for i in range(1, n_goods):
            f.write(f'v_{i}*x_1{i}*fn_{i}+v_plus{i}*x_2{i}plus*fn_plus{i}=v_0*x_10*fn_0+v_plus0*x_20plus*fn_plus0;')
            f.write('\n')
        
        f.write(f'\n')
        for i in range(1, n_goods):
            f.write(f'v_{i}*fs_{i}=v_0*fs_0;')
            f.write('\n')
        
        f.write(f'\n')
        f.write(f'bet*(fs_0plus/exp(k)+z_2plus/z_1plus)=v_0/v_plus0/z_1;\n')
        
        f.write(f'\n')
        f.write(f'exp(k)=(1-delt)*exp(k(-1))+i-psi_k/2*(i/exp(k(-1))-delt)^2*exp(k(-1));\n')
                
        f.write(f'\n')
        
        
        f.write(f'\n')
        f.write(f'exp(c_0)=exp(a_0)*exp_n_0^(1-alf)*q_0^(1-alf)*exp_s_0^alf*exp(k(-1))^alf-i;\n')
        for i in range(1, n_goods):
            f.write(f'exp(c_{i})=exp(a_{i})*exp(n_{i})^(1-alf)*q_{i}^(1-alf)*exp(s_{i})^alf*exp(k(-1))^alf')
            if i==0:
                f.write(f'-i')
            f.write(';\n')
                       
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'a_{i}=rho_a_{i}*a_{i}(-1)+a_{i}_shock;\n')
                  
        f.write(f'\n')  
       
        
        
        
        f.write('\n')
        f.write(f'end;\n\n\n')        
# Call the function
write_to_file('t_model.m')
```
