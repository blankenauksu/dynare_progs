```python
n_goods=3

def write_to_file(filename):
    with open(filename, 'w') as f:
        f.write('model;\n')
        
        f.write('# ph=(\n')
        for i in range(0, n_goods):
            f.write(f'om_{i}*exp(c_{i})^((eps-1)/eps)')
            if i < n_goods-1:
                f.write(f'+')
            f.write('\n')    
        f.write(')^((1-tht*eps)/(eps-1));\n')
        f.write(f'\n')
        f.write('# ph_plus=(\n')
        for i in range(0, n_goods):
            f.write(f'om_{i}*exp(c_{i}(+1))^((eps-1)/eps)')
            if i < n_goods-1:
                f.write(f'+')
            f.write('\n')    
        f.write(')^((1-tht*eps)/(eps-1));\n')
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# v_{i}=om_{i}*exp(c_{i})^(-1/eps)*ph;\n')
       
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# v_plus{i}=om_{i}*exp(c_{i}(+1))^(-1/eps)*ph_plus;\n')
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# q_{i}=(1-psi_n/2*(exp(n_{i})/exp(n_{i}(-1))-1)^2);\n')
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# q_plus{i}=(1-psi_n/2*(exp(n_{i}(+1))/exp(n_{i})-1)^2);\n')
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# fn_{i}=exp(a_{i})*(1-alf)*(exp(s_{i})*exp(k(-1))/exp(n_{i})/q_{i})^alf;\n')   
            
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# fn_plus{i}=exp(a_{i}(+1))*(1-alf)*(exp(s_{i}(+1))*exp(k)/exp(n_{i}(+1))/q_plus{i})^alf;\n')       
         

        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# fs_{i}=exp(a_{i})*alf*exp((n_{i})/exp(s_{i})*q_{i})^(1-alf)*exp(k(-1))^alf;\n')
        f.write(f'# fs_0plus=exp(a_0(+1))*alf*(exp(n_0(+1))/exp(s_0(+1))*q_plus0)^(1-alf)*exp(k)^alf;\n')
         
        
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# x_1{i}=(q_{i}-exp(n_{i})*psi_n/exp(n_{i}(-1))*(exp(n_{i})/exp(n_{i}(-1))-1));\n')
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# x_2{i}plus=psi_n*(exp(n_{i}(+1))/exp(n_{i})-1)*(exp(n_{i}(+1))/exp(n_{i}))^2;\n')
            
        f.write(f'\n')
        f.write(f'# z_1=1-psi_k*(exp(i)/exp(k(-1))-delt);\n')
        f.write(f'# z_1plus=1-psi_k*exp((i(+1))/exp(k)-delt);\n')       
            
        f.write(f'\n')
        f.write(f'# z_2plus=1-delt-psi_k/2*exp((i(+1))/exp(k)-delt)^2+psi_k*exp(i(+1))/exp(k)*(exp(i(+1))/exp(k)-delt);\n') 
        
        f.write(f'\n') 
        for i in range(0, n_goods):
            f.write(f'exp(k_{i})=exp(s_{i})/exp(n_{i})*exp(k(-1));\n')
        
        
        
        
        
        
        f.write(f'\n')
        for i in range(1, n_goods):
            f.write(f'v_{i}*x_1{i}*fn_{i}+v_plus{i}*x_2{i}plus*fn_plus{i}=v_0*x_10*fn_0+v_plus0*x_20plus*fn_plus0;')
            f.write('\n')
        
        f.write(f'\n')
        for i in range(1, n_goods):
            f.write(f'v_{i}*fs_{i}=v_0*fs_0;')
            f.write('\n')
        
        f.write(f'\n')
        f.write(f'bet*(fs_0plus/exp(k)+z_2plus/z_1plus)=v_0/v_plus0/z_1;\n')
        
        f.write(f'\n')
        f.write(f'exp(k)=(1-delt)*exp(k(-1))+exp(i)-psi_k/2*(exp(i)/exp(k(-1))-delt)^2*exp(k(-1));\n')
                
        f.write(f'\n')
        f.write('1=')
        for i in range(0, n_goods):
            f.write(f'exp(n_{i})')
            if i < n_goods-1:
                f.write(f'+')            
        f.write(';\n')
        
        f.write(f'\n')
        f.write('1=')
        for i in range(0, n_goods):
            f.write(f'exp(s_{i})')
            if i < n_goods-1:
                f.write(f'+')            
        f.write(';\n')
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'exp(c_{i})=exp(a_{i})*exp(n_{i})^(1-alf)*q_{i}^(1-alf)*exp(s_{i})^alf*exp(k(-1))^alf')
            if i==0:
                f.write(f'-exp(i)')
            f.write(';\n')
                       
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'a_{i}=rho_a_{i}*a_{i}(-1)+a_{i}_shock;\n')
                  
        f.write(f'\n')  
        
        f.write(f'exp(y_0)=exp(c_0)+exp(i);\n')
        f.write('\n')
        
        
        for i in range(0, n_goods):
            f.write(f'exp(p_{i})=om_{i}/om_0*(exp(c_{i})/exp(c_0))^(-1/eps);\n')
                  
        f.write('\n')
              
        f.write('\n')
      
        f.write(f'exp(y)=exp(i)+exp(c_0)*exp(p_0)+exp(c_1)*exp(p_1)+exp(c_2)*exp(p_2);\n')
        
       
        f.write(f'exp(r)=1;\n')
        #f.write(f'r=1-delt+alf*k^(alf-1)*(s_0/n_0)^alf;\n')
        f.write('\n')
        
        f.write(f'tot=(1/exp(r)*exp(y)-exp(y(-1)))/exp(y(-1));\n')
        f.write('\n')
                
        f.write(f'sc=(\n')
        f.write(f'(1/exp(r)*exp(y_0)*exp(p_0)/exp(n_0)-exp(y_0(-1))*exp(p_0(-1))/exp(n_0(-1)))*exp(n_0)+\n')
        for i in range(1, n_goods):
            f.write(f'(1/exp(r)*exp(c_{i})*exp(p_{i})/exp(n_{i})-exp(c_{i}(-1))*exp(p_{i}(-1))/exp(n_{i}(-1)))*exp(n_{i})')
            if i < n_goods-1:
                f.write(f'+\n')
            else:
             f.write(f')\n')  
            
        f.write('/exp(y(-1));\n')
        f.write('\n')
         
       
     
        f.write(f'win=(\n')
        f.write(f'(exp(n_0)-exp(n_0(-1)))*exp(y_0(-1))*exp(p_0(-1))/exp(n_0(-1))+\n')
        for i in range(1, n_goods):
            f.write(f'(exp(n_{i})-exp(n_{i}(-1)))*exp(c_{i}(-1))*exp(p_{i}(-1))/exp(n_{i}(-1))')
            if i < n_goods-1:
                f.write(f'+\n')
            else:
                f.write(f')\n')
        f.write('/exp(y(-1));\n')
        f.write('\n')           

        #f.write('ex_0=exp(a_0)*n_0^(1-alf)*q_0^(1-alf)*s_0^alf*k(-1)^alf;\n')
        #f.write('ex_1=exp(a_0)*n_0^(1-alf)*q_0^(1-alf);\n')
        #f.write('ex_2=s_0^alf*k(-1)^alf;\n')
        
        f.write('ex_0=0;\n')
        f.write('ex_1=0;\n')
        f.write('ex_2=0;\n')
        f.write('exb_0=0;\n')
        f.write('exb_1=0;\n')
        f.write('exb_2=0;\n')
        
        
        
        
        
        #f.write('ex_0=y_0(-1)*p_0(-1)/n_0(-1);\n')
        #f.write('ex_1=c_1(-1)*p_1(-1)/n_1(-1);\n')
        #f.write('ex_2=c_2(-1)*p_2(-1)/n_2(-1);\n')
        #f.write('exb_0=(n_0-n_0(-1));\n')
        #f.write('exb_1=(n_1-n_1(-1));\n')
        #f.write('exb_0=(s_1(-1)/n_1(-1)-s_0(-1)/n_0(-1));\n')
        #f.write('exb_1=(n_1+n_2-n_1(-1)-n_2(-1));\n')
        #f.write('exb_2=(s_1(-1)/n_1(-1)-s_0(-1)/n_0(-1))*(n_1+n_2-n_1(-1)-n_2(-1));\n')
        #f.write('exb_2=(n_2-n_2(-2));\n')
        #f.write('win=ex_0*exb_0+ex_1*exb_1+ex_2*exb_2;\n')
 
        
        
        
        
        f.write('\n')
        f.write(f'end;\n\n\n')        
# Call the function
write_to_file('t_model.m')
```
