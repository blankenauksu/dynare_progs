```python
n_goods=3

def write_dict_to_file(filename):
    with open(filename, 'w') as f:
        f.write('model;\n')
        f.write('# ph=(\n')
        for i in range(0, n_goods):
            f.write(f'om_{i}*c_{i}^((eps-1)/eps)')
            if i < n_goods-1:
                f.write(f'+')
            f.write('\n')    
        f.write(')^((1-tht*eps)/(eps-1));\n')

        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# v_{i}=om_{i}*c_{i}^(-1/eps)*ph;\n')
        f.write(f'# v_0lead=om_0*c_0(+1)^(-1/eps)*ph;\n')    

        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'# fk_{i}=exp(a_{i})*n_{i}*alf*k(-1)^(alf-1);\n')
        f.write(f'# fk_0lead=exp(a_0(+1))*n_0(+1)*alf*k^(alf-1);\n')
        
        #f.write(f'\n')
        #for i in range(0, n_goods):
        #    f.write(f'# fn_{i}=a_{i}*(1-alf)*k^(alf);\n')
        
        f.write(f'\n')
        f.write(f'# lamk=v_0*fk_0/n_0;\n')
        
        
        f.write(f'\n')
        for i in range(1, n_goods):
            f.write(f'v_{i}*fk_{i}=n_{i}*lamk;\n')
        
        f.write(f'\n')
        f.write(f'bet*(fk_0lead+1-delt)=v_0/v_0lead;\n')
        
        
        f.write(f'\n')
        f.write('1=')
        for i in range(0, n_goods):
            f.write(f'n_{i}')
            if i < n_goods-1:
                f.write(f'+')            
        f.write(';\n')
        
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'c_{i}=exp(a_{i})*n_{i}*k(-1)^alf')
            if i==0:
                f.write(f'+(1-delt)*k(-1)-k')
            f.write(';\n')
                  
        f.write(f'\n')
        for i in range(0, n_goods):
            f.write(f'a_{i}=rho_a_{i}*a_{i}(-1)+a_{i}_shock;\n')
            
            
        f.write(f'\n')  
        f.write(f'end;\n\n\n')        
# Call the function
write_dict_to_file('t_ss_model.m')
```
