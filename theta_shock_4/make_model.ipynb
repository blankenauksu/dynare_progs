```python


def write_to_file(filename):
    with open(filename, 'w') as f:
        f.write('model;\n')
        f.write(f'\n')
       
        f.write(f'\n')
        f.write(f'# x0=1;\n')
        f.write('# X=\n')
        for i in range(0, n_goods):
            f.write(f'(exp(k(-1))*exp(s0)/exp(n0))^((theta0-theta{i})*(1-eps))*(omega{i}/omega0)^eps* \n')
            f.write(f'(exp(a0)^(1-theta0)/exp(a{i})^(1-theta{i})*(theta0/theta{i})^theta{i}*((1-theta0)/(1-theta{i}))^(1-theta{i}))^(1-eps)\n')
            if i < n_goods-1:
                f.write(f'+')               
        f.write(';\n')
        f.write(f'\n')
        
        f.write('# Z=\n')
        for i in range(0, n_goods):
            f.write(f'theta{i}*\n')
            f.write(f'exp((k(-1))*exp(s0)/exp(n0))^((theta0-theta{i})*(1-eps))*(omega{i}/omega0)^eps* \n')
            f.write(f'(exp(a0)^(1-theta0)/exp(a{i})^(1-theta{i})*(theta0/theta{i})^theta{i}*((1-theta0)/(1-theta{i}))^(1-theta{i}))^(1-eps)\n')
            if i < n_goods-1:
                f.write(f'+')            
        f.write(';\n')
        f.write(f'\n')
        
        f.write(f'#Q=(Z-theta0*X)/X;\n')
        

        
        f.write(f'\n')
        f.write(f'exp(c(+1))=exp(c)*bet*(theta0(+1)*(exp(k)*exp(s0(+1))/exp(n0(+1)))^(theta0(+1)-1)+1-delt);\n')
        
        f.write(f'exp(n0)=(1-theta0)*(1-exp(c)/exp(y)*(1-x0/X))/(1-theta0-exp(c)/exp(y)*Q);\n')
        f.write(f'exp(s0)=theta0*(1-exp(c)/exp(y)*(1-x0/X))/(theta0+exp(c)/exp(y)*Q);\n')
        f.write(f'exp(c)/X=(exp(s0)*exp(k(-1)))^theta0*(exp(a0)*exp(n0))^(1-theta0)-i;')
       
        f.write(f'\n')
        f.write(f'exp(y)=(1-theta0)*exp(a0)^(1-theta0)*exp(k(-1))^theta0*(exp(s0)/exp(n0))^theta0/(1-theta0-exp(c)/exp(y)*Q);\n')
        f.write(f'\n')
    
        f.write(f'i=exp(k)*(1+gamman)*(1+gamma0)-(1-delt)*exp(k(-1));\n')
        f.write(f'\n')
                   
        for i in range(0, n_goods):
            f.write(f'a{i}=a_rho{i}*a{i}(-1)+a{i}shock;\n')
            
        f.write(f'\n')     
        for i in range(0, n_goods):
            f.write(f'theta{i}=2*((1-theta_rho{i})*theta_mean{i}+theta_rho{i}*theta{i}(-1))/(1+vp*exp(theta{i}shock));\n')
        f.write(f'\n')
        
        f.write(f'end;\n')  
        
           
# Call the function
write_to_file('t_model.m')              
    
```
